```python
%matplotlib inline
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
mergedf = pd.read_csv('metadata_otu_merged.csv')
mergedf.head(1)
```

```python
mergedf = mergedf.drop("Unnamed: 0", axis=1)
```

```python
print(mergedf.shape)
```

```python
print(mergedf.iloc[:,[1085]])
```

```python
X = mergedf.iloc[:, 52:1085]
y = mergedf["sample_type"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
# y_scaler = StandardScaler().fit(y_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
# y_train_scaled = y_scaler.transform(y_train)
# y_test_scaled = y_scaler.transform(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
y_train.shape
```

```python
model = classifier.fit(X_train, y_train.ravel())
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test.ravel()}).reset_index(drop=True).head(10)
```

```python
print("_____saliva, skin, stool")
confusion_matrix(y_test, predictions, labels=["saliva","skin","stool"])
```
